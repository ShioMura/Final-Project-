```python
import numpy as np
import sep
import matplotlib.pyplot as plt
import astropy

from astropy.io import fits
from matplotlib import rcParams

from astropy.visualization import make_lupton_rgb
from astropy.utils.data import get_pkg_data_filename
#http://docs.astropy.org/en/v3.0.2/visualization/lupton_rgb.html

%matplotlib inline

rcParams['figure.figsize'] = [10., 8.]

save_path = '/Users/labuser/Desktop/'
```

```python
R_name = get_pkg_data_filename("/Users/labuser/Downloads/hlsp_hudf12_hst_wfc3ir_udfmain_f160w_v1.0_drz.fits")
G_name = get_pkg_data_filename("/Users/labuser/Downloads/hlsp_hudf12_hst_wfc3ir_udfmain_f125w_v1.0_drz.fits")
B_name = get_pkg_data_filename("/Users/labuser/Downloads/hlsp_hudf12_hst_wfc3ir_udfmain_f105w_v1.0_drz.fits")

#read the data using fits.getdata()
R = fits.open(R_name)[0].data
G = fits.open(G_name)[0].data
B = fits.open(B_name)[0].data
```

```python
RGB_default = make_lupton_rgb(R, G, B, Q=10, stretch=0.05, filename = "HUDF_RGB.png")

#show the stacked image
plt.imshow(RGB_default, origin='lower')
```
